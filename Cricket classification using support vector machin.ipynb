```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("cricket.csv")
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
for i in ds.columns:
  if (ds[i].dtypes != 'object'):
    q1 = ds[i].quantile(0.25)
    q2 = ds[i].quantile(0.75)
    IQR = q2 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q2 + 1.5 * IQR
    outliers = ds[(ds[i] < lower_bound) | (ds[i] > upper_bound)]
    if (len(outliers) > 0):
      imp = SimpleImputer(strategy = 'median')
      ds[[i]] = imp.fit_transform(ds[[i]])
    else:
      imp = SimpleImputer(strategy = 'mean')
      ds[[i]] = imp.fit_transform(ds[[i]])
  else:
    imp = SimpleImputer(strategy = 'most_frequent')
    ds[[i]] = imp.fit_transform(ds[[i]])
ds
```

```python
ds.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in ds.columns:
  if ds[i].dtypes=="object":
    ds[i]=ds[[i]].apply(le.fit_transform)
ds
```

```python
x=ds.drop("Inns",axis=1)
y=ds["Inns"]
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.svm import SVC
clf=SVC(kernel="rbf")
```

```python
clf.fit(x_train,y_train)
```

```python
y_pret=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pret,y_test)
```
